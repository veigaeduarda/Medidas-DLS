```python
import pandas as pd
import numpy as np

df60 = pd.read_csv('SD60 - correlograma.txt',names=["x","y"])
df30 = pd.read_csv('SD30- correlogram.txt',names=["x","y"])
df45 =pd.read_csv('SD45 - correlogram.txt',names=["x","y"])
```

```python
df60["x"] = np.log10(df60["x"])
df30["x"] = np.log10(df30["x"])
df45["x"] = np.log10(df45["x"])
```

```python
import matplotlib.pyplot as plt 


fig, axs = plt.subplots(3, sharex=True)



axs[0].plot(df60["x"], df60["y"])
axs[1].plot(df45["x"], df45["y"])
axs[2].plot(df30["x"], df30["y"])

fig.add_subplot(111, frame_on=False)

plt.tick_params(labelcolor="none", bottom=False, left=False)


plt.xlabel("Tempo")

plt.ylabel("Correlograma")

plt.savefig("correlograma")

plt.show()
```
